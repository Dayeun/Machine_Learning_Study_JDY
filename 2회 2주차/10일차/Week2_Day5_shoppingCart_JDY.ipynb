```python
class Apriori_algorithm():
    #
    def support(self, sparse_matrix, item_name):
        return sparse_matrix[item_name].mean()
    
    #
    def confidence(self, item_name1, item_name2):
        #return (sparse_matrix[item_name1]==True & (sparse_matrix[item_name1] == sparse_matrix[item_name2])).mean()
        return ((sparse_matrix[item_name1]==True) & (sparse_matrix[item_name2]==True)).mean()
        
    #
    def lift(self, item_name1, item_name2):
        confidence(item_name1, item_name2) / support(item_name1) * support(item_name2)
        
    
#     def shopping_analysis(sparse_matrix, ,Min_support)
```

```python
((sparse_matrix['all- purpose']==True) & (sparse_matrix['butter']==True)).mean()
```

```python
(((sparse_matrix['all- purpose']==True) & (sparse_matrix['butter']==True)).mean() / sparse_matrix['all- purpose'].mean() * sparse_matrix['butter']).mean()
```

```python
#   
Min_support = 0.31

Apriori = Apriori_algorithm()
support_list = [Apriori.support(sparse_matrix, i) for i in sparse_matrix.columns]
support_dict = dict(zip(sparse_matrix.columns, support_list))
support_dict
filter_support_list = [i for i in support_list if (i >= Min_support)]  #    : 0.38
filter_support_list
filter_support_dict = {k: v for k, v in support_dict.items() if v >= Min_support}
print('   ')
print(' : ')
(filter_support_dict)
```

```python
from itertools import combinations
comb = list(combinations(filter_support_dict, 2))
comb_confidence = [Apriori.confidence(i[0],i[1]) for i in comb]
comb_confidence_dict = dict(zip(comb, comb_confidence))
filter_confidence_dict = {k: v for k, v in comb_confidence_dict.items() if v >= Min_support}
print('   ')
print('(\'1\', \'2\') : ( )')
filter_confidence_dict
sorted(filter_confidence_dict.items(), key=lambda kv: kv[1], reverse=True)
```
