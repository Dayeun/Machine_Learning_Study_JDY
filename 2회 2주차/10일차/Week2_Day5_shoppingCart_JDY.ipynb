```python
# 모든 물품 목록 추출

#dataframe의 모든 element를 1차원 행렬로 바꾸기.
all_item=set(df.values.flatten())
all_item

# None, 공백만 있는 값 제거.
all_item.remove(None)
all_item.remove(' ')
all_item = list(all_item)  # 연산을 위해 set->list 변환.

#string 앞뒤에 있는 공백 없애기.
for i in range(len(all_item)):
    all_item[i] = all_item[i].strip()
    
#최종 '모든 물품 목록' 리스트.
all_item = list(set(all_item))
all_item.sort()
```

```python
# 인당 쇼핑카트 리스트

temp_cart = []
for i in range(len(df)):
    temp_cart_i = list(set(df.iloc[i,:]))
    if None in temp_cart_i:
        temp_cart_i.remove(None)
    if ' ' in temp_cart_i:
        temp_cart_i.remove(' ')
        
    #string 앞뒤에 있는 공백 없애기.
    for i in range(len(temp_cart_i)):
        temp_cart_i[i] = temp_cart_i[i].strip()
    temp_cart_i.sort()
        
    temp_cart.append(temp_cart_i)
temp_cart
```

```python
# 모든 고객의 쇼핑카트 리스트 물품 비교
shopping_cart = [[all_item[i] in temp_cart[j] for i in range(len(all_item))] for j in range(len(temp_cart))]
shopping_cart
```

```python
sparse_matrix = pd.DataFrame(shopping_cart, columns=all_item)
sparse_matrix
```

```python
sparse_matrix.dtypes
```

```python
class Apriori_algorithm():
    #지지도
    def support(self, sparse_matrix, item_name):
        return sparse_matrix[item_name].mean()
    
    #신뢰도
    def confidence(self, item_name1, item_name2):
        #return (sparse_matrix[item_name1]==True & (sparse_matrix[item_name1] == sparse_matrix[item_name2])).mean()
        return ((sparse_matrix[item_name1]==True) & (sparse_matrix[item_name2]==True)).mean()
        
    #향상도
    def lift(self, item_name1, item_name2):
        confidence(item_name1, item_name2) / support(item_name1) * support(item_name2)
        
    
#     def shopping_analysis(sparse_matrix, ,Min_support)
```

```python
((sparse_matrix['all- purpose']==True) & (sparse_matrix['butter']==True)).mean()
```

```python
(((sparse_matrix['all- purpose']==True) & (sparse_matrix['butter']==True)).mean() / sparse_matrix['all- purpose'].mean() * sparse_matrix['butter']).mean()
```

```python
# 최소 지지도 설정
Min_support = 0.31

Apriori = Apriori_algorithm()
support_list = [Apriori.support(sparse_matrix, i) for i in sparse_matrix.columns]
support_dict = dict(zip(sparse_matrix.columns, support_list))
support_dict
filter_support_list = [i for i in support_list if (i >= Min_support)]  # 최소 지지도 설정 : 0.38
filter_support_list
filter_support_dict = {k: v for k, v in support_dict.items() if v >= Min_support}
print('최소 지지도를 넘는 상품들')
print('상품명 : 지지도')
(filter_support_dict)
```

```python
from itertools import combinations
comb = list(combinations(filter_support_dict, 2))
comb_confidence = [Apriori.confidence(i[0],i[1]) for i in comb]
comb_confidence_dict = dict(zip(comb, comb_confidence))
filter_confidence_dict = {k: v for k, v in comb_confidence_dict.items() if v >= Min_support}
print('두 묶음 상품의 지지도')
print('(\'상품명1\', \'상품명2\') : 지지도(내림차순 정렬)')
filter_confidence_dict
sorted(filter_confidence_dict.items(), key=lambda kv: kv[1], reverse=True)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# 참고 코드 : https://stackoverflow.com/questions/33711107/create-heatmap-in-python-matplotlib-with-x-and-y-labels-from-dict-with-tuplefl
ser = pd.Series(list(filter_confidence_dict.values()),
                  index=pd.MultiIndex.from_tuples(filter_confidence_dict.keys()))
df = ser.unstack()
df.shape
```

```python
# Min_support = 0.31 일 때,

plt.figure(figsize=(10, 10))
sns.heatmap(df,square=True, annot=True)
plt.show()
plt.close()
```
