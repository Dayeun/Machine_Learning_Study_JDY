# Naive-Bayes Model - SPAM filter_JDY

```python
import pandas as pd
```

```python
df_raw = pd.read_csv('sms_spam.csv', encoding='iso-8859-1')
df_raw
```

### 숫자 제거

```python
#참고 사이트: https://stackoverflow.com/questions/45600662/pandas-remove-numbers-from-start-of-string-in-series?rq=1
#정수 혹은 실수를 제거.
df_delNum = df_raw.copy()
df_delNum['text']=df_delNum['text'].str.replace('\d|\f','')
df_delNum
```

### text 소문자화

```python
# 참고 사이트 : https://stackoverflow.com/questions/22245171/how-to-lowercase-a-python-dataframe-string-column-if-it-has-missing-values
df_lower = df_delNum.copy()
df_lower['text']=df_lower['text'].str.lower()
df_lower
```

### 불용어 제거 방법 둘

```python
# 불용어 제거 방법 1
df_delStopword_1 = df_lower.copy()
df_delStopword_1['text'] = df_delStopword_1['text'].str.replace('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]',' ')
df_delStopword_1
```

```python
# 참고 사이트 : https://stackoverrun.com/ko/q/12983392
# 불용어 제거 방법 2 
df_delStopword_2 = df_lower.copy()
df_delStopword_2 = df_delStopword_2.replace('[^\w\s]',' ',regex=True)
df_delStopword_2
```

### 전처리 결과 확인

```python
print('[19] 전처리 전:\n', df_raw['text'][19])
print('[19] 전처리 후:\n', df_delStopword_2['text'][19])
```

```python
print('[22] 전처리 전: ', df_raw['text'][22])
print('[22] 전처리 후: ', df_delStopword_2['text'][22])
```

### type: 'spam',  'ham' → 0, 1

```python
df = df_delStopword_2.copy()
df['type'] = df.type.map({'spam':0, 'ham':1})
print(df.dtypes)
df.head()
```

```python
# 스팸인 글들을 모아 보기
df[df['type']==0]
```

## SPAM 분류

### X, Y 설정

```python
# 참고 사이트: https://wikidocs.net/24869

# CountVectorizer : 문서 집합에서 단어 토큰을 생성하고 각 단어의 수를 세어 BOW 인코딩한 벡터를 만든다.
# CountVectorizer 참고 : https://datascienceschool.net/view-notebook/3e7aadbf88ed4f0d87a76f9ddc925d69/
from sklearn.feature_extraction.text import CountVectorizer

# n-gram 관련 참고 : https://datascienceschool.net/view-notebook/a0c848e1e2d343d685e6077c35c4203b/
count_ngrams = CountVectorizer(
    ngram_range=(1,3), analyzer="word", binary="False", stop_words="english")
X = count_ngrams.fit_transform(df['text'])
print(X.toarray()) 

print(count_ngrams.get_feature_names()[:5])

import numpy as np

Y = [1 if data == True else 0 for data in df['type']]
Y = np.array(Y)
print(Y)
```

### X, Y train / test data 나누고 베르누이 나이브 베이즈 모델 돌리면서 교차검증

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.model_selection import KFold
from sklearn import linear_model, datasets
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score

##kf = KFold(Y.shape[0], n_folds=5,shuffle=True, )
# K-Folds cross-validator
kf = KFold(n_splits=5,shuffle=True, )
clf = BernoulliNB()   # 0.91
# clf = MultinomialNB()   # 0.95

##for i, (train_index, test_index) in enumerate(kf):
for train_index, test_index in kf.split(X):    
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]

    # ravel() : Return a flattened array.
    clf.fit(X_train, y_train.ravel())

    # accuracy_score(y_true, y_pred, normalize=True, sample_weight=None) : Accuracy classification score.
    print('accuracy:',accuracy_score(y_pred = clf.predict(X_test), y_true = y_test))
```

```python
df['predict'] = clf.predict(X)  # BernoulliNB()
df
```
