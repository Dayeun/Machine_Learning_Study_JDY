### 두 분류 평점 등급 예측 정확도 : 87%      [80~90 : 0, 90~100 : 1] 
### 네 분류 평점 등급 예측 정확도 : 83%      [80~85 : 0, 85~90 : 1, 90~95 : 2, 95~100 : 3]

```python
import pandas as pd
```

```python
df = pd.read_csv('winemag-data_first150k.csv', encoding='UTF-8', index_col=0)
df
```

```python
df.isna().sum()
```

```python
df['points'].describe()
```

```python
# 80~90 90~100
df['rank']=df.points.apply(lambda x : 0 if (x >=80 and x<90) else 1)

# 80~85  85~90  90~95  95~100
# 참고 사이트: https://stackoverrun.com/ko/q/12356081
# 가독성 갖다 버렸지만 빠르다.
df['rank2']=df.points.apply(lambda x : 
                           0 if (x >=80 and x<85) else (
                               1 if (x >=85 and x<90) else (
                                   2 if (x >=90 and x<95) 
                                   else 3 )))
df
```

```python
# description 숫자 제거
df['description']=df['description'].str.replace('\d|\f','')
# description 소문자화
df['description']=df['description'].str.lower()
# description 불용어 제거
df['description'] = df['description'].replace('[^\w\s]',' ',regex=True)

df
```

### 두 분류 rank

```python
# 참고 사이트: https://wikidocs.net/24869

# CountVectorizer : 문서 집합에서 단어 토큰을 생성하고 각 단어의 수를 세어 BOW 인코딩한 벡터를 만든다.
# CountVectorizer 참고 : https://datascienceschool.net/view-notebook/3e7aadbf88ed4f0d87a76f9ddc925d69/
from sklearn.feature_extraction.text import CountVectorizer

# n-gram 관련 참고 : https://datascienceschool.net/view-notebook/a0c848e1e2d343d685e6077c35c4203b/
# binary="False" 왜 큰따옴표 안에 False를 넣었을까.. 이래도 False로 먹히나??
count_ngrams = CountVectorizer(
    ngram_range=(1,3), analyzer="word", binary="False", stop_words="english")
X = count_ngrams.fit_transform(df['description'])
#print(X.toarray()) 

print(count_ngrams.get_feature_names()[:5])

import numpy as np

Y = [1 if data == True else 0 for data in df['rank']]
Y = np.array(Y)
print(Y)
```

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.model_selection import KFold
from sklearn import linear_model, datasets
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score

##kf = KFold(Y.shape[0], n_folds=5,shuffle=True, )
# K-Folds cross-validator
kf = KFold(n_splits=5,shuffle=True, )
clf = BernoulliNB()   
# clf = MultinomialNB()   

##for i, (train_index, test_index) in enumerate(kf):
for train_index, test_index in kf.split(X):    
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]

    # ravel() : Return a flattened array.
    clf.fit(X_train, y_train.ravel())

    # accuracy_score(y_true, y_pred, normalize=True, sample_weight=None) : Accuracy classification score.
    print('accuracy:',accuracy_score(y_pred = clf.predict(X_test), y_true = y_test))
```

```python
df['predict_rank'] = clf.predict(X)  # BernoulliNB()
df
```

### 네 분류 rank

```python
# 참고 사이트: https://wikidocs.net/24869

# CountVectorizer: 문서 집합에서 단어 토큰을 생성하고 각 단어의 수를 세어 'BOW 인코딩'한 벡터를 만든다.
# 참고 사이트 : https://datascienceschool.net/view-notebook/3e7aadbf88ed4f0d87a76f9ddc925d69/
from sklearn.feature_extraction.text import CountVectorizer

# n-gram 관련 참고 : https://datascienceschool.net/view-notebook/a0c848e1e2d343d685e6077c35c4203b/
count_ngrams = CountVectorizer(
    ngram_range=(1,3), analyzer="word", binary="False", stop_words="english")
X = count_ngrams.fit_transform(df['description'])
#print(X.toarray()) 

print(count_ngrams.get_feature_names()[:5])

import numpy as np

Y = [data for data in df['rank2']]
Y = np.array(Y)
print(Y)
```

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.model_selection import KFold
from sklearn import linear_model, datasets
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score

##kf = KFold(Y.shape[0], n_folds=5,shuffle=True, )
# K-Folds cross-validator
kf = KFold(n_splits=5,shuffle=True, )
clf = MultinomialNB()   # 83%
# clf = BernoulliNB()   # 60%

##for i, (train_index, test_index) in enumerate(kf):
for train_index, test_index in kf.split(X):    
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]

    # ravel() : Return a flattened array.
    clf.fit(X_train, y_train.ravel())

    # accuracy_score(y_true, y_pred, normalize=True, sample_weight=None) : Accuracy classification score.
    print('accuracy:',accuracy_score(y_pred = clf.predict(X_test), y_true = y_test))
```

```python
df['predict_rank2'] = clf.predict(X)  # MultinomialNB()
df
```
